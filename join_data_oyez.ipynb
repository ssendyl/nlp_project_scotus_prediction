```python
def fix_row(row):
    if row['docket_number'] and row['term']:
        row['docketId'] = str(row['term']) + '-' + str(row['docket_number'])
    return row
```

```python
df_2 = facts_df.apply(fix_row,axis=1)
```

```python
RESULT = pd.merge(df,df_2,how='left',on='docketId')
```
